```python
def total_words(file_name):
    '''
        Function to print the number of tokens. 
        To check if the total number of words are more than a million (100,000+)
    parameter: file_name:- name of the corpus  
    '''
    filename = file_name

    with open(filename, 'r') as file:
        lines_in_file = file.read()

        nltk_tokens = nltk.word_tokenize(lines_in_file)
        print ("\n Number of Words: " , len(nltk_tokens))
        return len(nltk_tokens)
```

```python
# calling the function
total_words('kiswa_corpus.txt')
```

```python
def tokens(file_name):
    '''
        Function to print the tokens in the corpus 
    parameter: file_name- name of the file
    '''
    
    filename = file_name

    with open(filename, 'r') as file:
        lines_in_file = file.read()

        nltk_tokens = nltk.word_tokenize(lines_in_file)
        return nltk_tokens
```

```python
tokens('kiswahili.txt')
```

```python
def sent_tokens(file_name):
    '''
        Function to split the kiswa corpus  
    parameter: file_name- name of the file
    '''
    
    filename = file_name

    with open(filename, 'r') as file:
        lines_in_file = file.read()
        sent = nltk.sent_tokenize(lines_in_file)
        length=len(sent)
        return sent,length
```

# Data processing 

```python
def clean_text(sentence):
    '''
        This function cleans the column
    '''
    sentence = sentence.lower()                #convert text to lower-case
    sentence = re.sub('[‘’“”…,]', '', sentence) # remove punctuation
    sentence = re.sub('[()]', '', sentence)  #remove parentheses
    sentence = word_tokenize(sentence)      # remove repeated characters (tanzaniaaaaaaaa to tanzania)
    
    return ' '.join(sentence)
```

```python
# converting the float into string. 
df = df.astype(str)
```

```python
# calling the clean_text function on news_heading_swa
df['news_heading_swa'] = df['news_heading_swa'].apply(clean_text) 
df['news_heading_eng'] = df['news_heading_eng'].apply(clean_text) 
df.head()
```

```python
# Splitting the original data into train and test on 90/ 10 split (train/test)
SEED = 2020
df_train= df.sample(frac=0.90,random_state=SEED)
df_test = df.drop(df_train.index)
len(df_train),len(df_test)
```
